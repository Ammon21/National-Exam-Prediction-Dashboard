# Machine Learning

```python
Science = df[df.Type == 'N']
```

```python
Science.Batch.value_counts()
```

```python
Science.info()
```

```python
dfx = Science.drop(['Geography', 'History', 'Name'], axis=1)
```

```python
dfx.info()
```

```python
dfx_2023 = dfx[dfx.Batch == 2023]
dfx_2022 = dfx[dfx.Batch == 2022]
```

```python
dfx_2022.info()
```

```python
dfx_2022
```

```python
reg = linear_model.LinearRegression()
reg.fit(dfx_2022[['Maths', 'English', 'Civics', 'Chemistry', 'Biology', 'Physics', 'GPA']], dfx_2022.Matrik)
```

```python
reg.coef_ 
```

```python
reg.intercept_
```
